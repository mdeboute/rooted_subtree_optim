```python
import os

def parse_instance(instance):
    with open(instance, 'r') as f:
        lines = f.readlines()
        opt = float(lines[-1].split(':')[1].strip())
    return opt

def parse_dir(path):
    filenames = []
    optimal_solutions = []
    dir = os.listdir(path)
    dir.sort()
    for f in dir:
        filenames.append(f)
        optimal_solutions.append(parse_instance(os.path.join(path, f)))
    return filenames, optimal_solutions
```

```python
filenames, optimal_solutions = parse_dir('./Instances')
```

```python
def parse_bilan(filename):
    times = []
    solutions = []
    with open(filename, 'r') as f:
        for line in f:
            times.append(float(line.split(';')[2].strip()))
            solutions.append(float(line.split(';')[4].strip()))
    return times, solutions
```

```python
times_model_1, solutions_model_1 = parse_bilan('./Solutions_model_1/bilan.csv')
times_model_2, solutions_model_2 = parse_bilan('./Solutions_model_2/bilan.csv')
```

```python
import numpy as np

print(np.allclose(optimal_solutions, solutions_model_1, rtol=1e-03, atol=1e-06, equal_nan=True))
```

```python
print(np.allclose(optimal_solutions, solutions_model_2, rtol=1e-03, atol=1e-06, equal_nan=True))
```

```python
import matplotlib.pyplot as plt

plt.plot([i for i in range(len(solutions_model_1))], times_model_1)
plt.xlabel('Index of instance')
plt.ylabel('Time (s)')
plt.title('Time to solve instances for model 1')
plt.show()
```
